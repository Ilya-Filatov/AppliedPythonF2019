### Step 3. Assign it to a variable called crime.

```python
crime = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv', sep = ',')
crime.head(10)
```

### Step 4. What is the type of the columns?

```python
crime.dtypes
```

##### Have you noticed that the type of Year is int64. But pandas has a different type to work with Time Series. Let's see it now.

### Step 5. Convert the type of the column Year to datetime64

```python
crime['Year'] = pd.to_datetime(crime['Year'], format = '%Y')
crime.head()
```

```python
crime.dtypes
```

### Step 6. Set the Year column as the index of the dataframe

```python
crime.index = pd.DatetimeIndex(crime['Year']).year
crime.drop(['Year'], axis = 1, inplace = True)
crime.sample(5)
```

### Step 7. Delete the Total column

```python
crime.drop(['Total'], axis = 1, inplace = True)
crime.sample(5)
```

### Step 8. Group the year by decades and sum the values

#### Pay attention to the Population column number, summing this column is a mistake

```python
df1 = crime.groupby(crime.index//10)['Violent'].count()
crime_gb = crime.groupby(crime.index//10).sum()
tmp = crime_gb['Population'] 
for i in range(len(df1)):
    tmp[tmp.index[i]] /= df1.iloc[i]
crime_gb['Population'] = tmp
crime_gb.index *= 10
crime_gb
```

### Step 9. What is the mos dangerous decade to live in the US?

```python
crime_gb['total'] = crime_gb.loc[:,'Violent':'Vehicle_Theft'].sum(axis=1)
crime_gb
```

```python
value = 0
for i in range(len(crime_gb)):
    k = crime_gb['total'].iloc[i] / crime_gb['Population'].iloc[i]
    if k > value:
        value = k
        res = crime_gb.index[i]
print('По количеству преступлений: ', crime_gb['total'].idxmax())
print('По количеству преступлений на душу населения: ', res)
```
